# Baseline Model

```python
# Was having problem to import but its the kernal after restart it work fine 
# from cuml.preprocessing import train_test_split
# from cuml.preprocessing.model_selection import train_test_split
from cuml.preprocessing import train_test_split
```

```python
# split data to train/test
X = df_ml.drop(['LOAN_ID','ORIG_UPB'], axis=1)
y = df_ml.ORIG_UPB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# BASELINE MODEL 
def baseline_model(n_preds, pred):
    # just predict the average
    return cudf.Series([pred for n in range(n_preds)])

# make baseline preds
baseline_preds = baseline_model(len(y_test), np.mean(y_train)).astype(np.float32)# had to fix preds type
```

```python
# evaluate model
from cuml.metrics import mean_squared_error
mean_squared_error(y_true=y_test,
                   y_pred=baseline_preds,
                   squared=False)
```

# Random Forest Regressor Model

```python
# create & fit model
from cuml.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()

rfr.fit(X_train, y_train)
```

```python
# model prediction
preds = rfr.predict(X_test)
```

```python
preds.tail(5)
```

```python
# evaluate model
from cuml.metrics import mean_squared_error
mean_squared_error(y_true=y_test,
                   y_pred=preds,
                   squared=False) # off by 22,696.613$
```
